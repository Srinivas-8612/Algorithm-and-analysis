```python
N = 12
DIS = float('inf')
def shortest_distance (graph):
#Initialize distance array with infinity
 distance = [DIS]* N
 distance [N-1]=0 # Destination (N-1) distance is O
#Process nodes from N-2 to 0
 for i in range (N-2, -1, -1):
#Check all possible nodes j that can go to (11)
   for j in range(i+1, N):
      if graph[i][j] == DIS:
        continue #No direct edge between i and j
#Update the distance of node i
      distance [i]=min (distance [i], graph[i][j]+ distance [j])
 return distance [0]
#Graph representation (adjacency matriz)
graph = [
[DIS, 9, 7, 3, 2, DIS, DIS, DIS, DIS, DIS, DIS, DIS],
[DIS, DIS, DIS, DIS, DIS, 4, 2, DIS, DIS, DIS, DIS, DIS],
[DIS, DIS, DIS, DIS, DIS, 2, 7, DIS, DIS, DIS, DIS, DIS],
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, 11, DIS, DIS, DIS, DIS],
# From node 5
[DIS, DIS, DIS, DIS, DIS, DIS, 11, 8, DIS, DIS, DIS, DIS],
#From node 6
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 6, 5, DIS, DIS],
#From node 7
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 4, 3, DIS, DIS],
#From node 8
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 5, 6, DIS],
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 4], #From node 9
[ DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 2], #From node 10
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 5], #From node 11
[DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS] #From node 12 (no outgoing edges)
]
#Calculate the shortest distance from start (node 0) to end (node N-1)
print("Distance from start:", shortest_distance (graph))
```
